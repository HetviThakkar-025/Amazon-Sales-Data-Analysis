**Summary of dataset**
- This dataset provides detailed information on 42,000+ Amazon electronics products, including sales, ratings, pricing trends, and sub-category distribution.

**Column Descriptions**



*   `title`: Complete name/title of the product
*   `rating`: Average customer rating out of 5
*   `number_of_reviews`: Total number of customer reviews
*   `bought_in_last_month`: Units purchased in the last month
*   `current/discounted_price`: Current price after discount
*   `price_on_variant`: a small attribute/variant shown on Amazon listings that helps identify product type or quantity.
*   `listed_price`: Original listed price before discount
*   `is_best_seller`: Indicates if the product is tagged as a Best Seller
*   `is_sponsored`: Whether the product is a Sponsored item or Organic
*   `is_couponed`: Special discounted coupons availability
*   `buy_box_availability`: BuyBox button availability on amazon search page like add to cart(nan values represent False)
*   `delivery_details`: Estimated delivery date
*   `sustainability_tags`: Eco-friendly and sustainability-related tags
*   `image_url`: Direct image link of the product
*   `product_url`: Official Amazon product page URL
*   `collected_at`: Date when the data was collected

















**Issues with the data:**

**1. Dirty Data:**
- Incorrect datatype assigned to `rating` and `listed_price` column (it should be 'int') => `validity`
- change name of col to shorter one and datatype to int for `current/discounted_price` => `validity`
- change datatype of `delivery_details` and `collected_at` to datetime => `validity`
- fill NaN values in rows other than 'X bought in past month' for `bought_in_last_month` column => `accuracy`
- For `listed_price` fill 'no discount' values with NaN and datatype to int => `accuracy & validity`

```python
# Copy dataset 
sales_df = sales.copy()
```

```python
sales_df.head(1)
```

```python
# First clean 'Rating' column
# Keep only numeric values and remove strings, also converting k to 1000 where value is > 1000
# Change datatype to float
sales_df['rating'] = sales_df['rating'].str.split(' ').str.get(0)
```

```python
sales_df['rating'] = sales_df['rating'].astype(float)
```

```python
sales_df.info()
```

```python
# Now 'listed_price' column
# Fill NaN values where value is No Discount & remove dollar sign ($)
# Convert datatype to float
sales_df['listed_price'] = sales_df['listed_price'].replace('No Discount', np.nan)
```

```python
sales_df['listed_price'].isnull().sum()
```

```python
sales_df['listed_price'] = sales_df['listed_price'].str.split('$').str.get(1).str.strip()
```

```python
sales_df['listed_price'] = sales_df['listed_price'].str.replace(',', '')
```

```python
sales_df['listed_price'] = sales_df['listed_price'].astype(float)
```

```python
# Rename column names
sales_df = sales_df.rename(columns={'current/discounted_price': 'discount', 'listed_price':'original_price'})
```

```python
# Remove commas in 'current/discounted_price' (discount) column
# Change datatype to float
sales_df['discount'] = sales_df['discount'].str.replace(',', '')
```

```python
sales_df['discount'] = sales_df['discount'].astype(float)
```

```python
sales_df.info()
```

```python
# Extract day and date from 'delivery_details' column 
# Change datatype to DateTime format
sales_df['delivery_details'] = sales_df['delivery_details'].str.split(' ').str[-2:].str.join(' ')
```

```python
# Replace with NaN where values is some text (not day or date)
import re

pattern = r'^[A-Za-z]{3} \d{1,2}$'
sales_df['delivery_details'] = sales_df['delivery_details'].where(sales_df['delivery_details'].str.match(pattern), np.nan)
```

```python
sales_df['delivery_details'] = pd.to_datetime(sales_df['delivery_details'], format='%b %d')
```

```python
# For sales data year 2025
sales_df['delivery_details'] = sales_df['delivery_details'].apply(
    lambda x: x.replace(year=2025) if pd.notnull(x) else pd.NaT
)
```

```python
# Change datatype to DateTime for data 'collected_at' column
sales_df['collected_at'] = sales_df['collected_at'].astype('datetime64[ns]')
```

```python
# Rename columns
sales_df = sales_df.rename(columns={'delivery_details': 'delivery_date', 'collected_at':'data_collected_at'})
```

```python
sales_df.info()
```

```python
# Replace with NaN in 'bought_in_last_month' column where value is some text which is not giving information about how many bought the product in last month
pattern = r'^[0-9.,]+[KkMm]?\+ bought in past month$'

sales_df['bought_in_last_month'] = sales_df['bought_in_last_month'].where(
    sales_df['bought_in_last_month'].str.match(pattern, na=False),
    np.nan
)
```

```python
sales_df['bought_in_last_month'] = sales_df['bought_in_last_month'].str.split('+').str.get(0)
```

```python
# Convert K to 1000 so to convert datatype from Object to float
K=1000

sales_df['bought_in_last_month'] = sales_df['bought_in_last_month'].apply(lambda x: float(x.replace('K','')) * 1000 if 'K' in str(x) else x).astype(float)
```

```python
sales_df.info()
```

```python
sales_df = sales_df.rename(columns={'number_of_reviews': 'reviews'})
```

```python
# Set product title as index
sales_df.set_index('title', drop=True, inplace=True)
```
